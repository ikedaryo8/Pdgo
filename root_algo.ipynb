```python
import folium
import openrouteservice
from branca.element import Figure
from openrouteservice import convert
import pandas as pd
import time

input_file_name = "C:\\Users\\Guren\\Downloads\\anaconda_py\\pd2.xlsx"

df = pd.read_excel(input_file_name)

key = "5b3ce3597851110001cf62483d4833f8cf424e4fa8397d2d2974a687"
client = openrouteservice.Client(key=key)
Line = []

def reverse_lat_long(list_of_lat_long):
    """緯度経度をひっくり返す"""
    return [(p[1], p[0]) for p in list_of_lat_long]

def directions(*arg):
    function_entry_data = []
    n = 0
    #停車するバス停をリストに追加
    for val in arg:
        function_entry_data.append(int(val))
    function_entry_data.append(int(32))
    #重複の削除
    set_data = function_entry_data
    #並び替え
    set_data.sort()
    return set_data


s  = input().split()

list = directions(*s)

 # foliumでサイズ(600, 400)の地図を描画
fig = Figure(width=600, height=400)
m = folium.Map(location=[36.5294566, 136.6145762], zoom_start=15)

for num in range(len(list) - 1):
    a = int(list[num])
    b = int(list[num + 1])

    p1 = []
    p2 = []

    for i,j in df.iterrows():
        if i == a:
            p1.append(j[1])
            p1.append(j[2])
        else:
            continue
    for i,j in df.iterrows():
        if i == b:
            p2.append(j[1])
            p2.append(j[2])
        else:
            continue


    p1r = tuple(reversed(p1))
    p2r = tuple(reversed(p2))

    # 経路計算 (Directions V2)
    routedict = client.directions((p1r, p2r))
    geom = routedict["routes"][0]["geometry"]
    decoded = convert.decode_polyline(geom)
    #time.sleep(0.2)

    coord = reverse_lat_long(decoded["coordinates"])

    # 位置情報をPolyLineで地図に追加
    folium.vector_layers.PolyLine(locations=coord).add_to(m)


for i, r in df.iterrows():
    folium.Marker(location=[r['緯度'],r['経度']], popup=r['地名']).add_to(m)
    
fig.add_child(m)

m.save("map5.html")
m
```
